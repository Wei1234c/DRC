# Fourier Transform

```python
%matplotlib notebook 

import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10
```

```python
fig = plt.figure(figsize = (10, 8))
ax = fig.gca(projection='3d')

freqs = (400, 1000)
n = 1001
x = np.linspace(0, 1, n)
x = np.linspace(0, 1/400 * 5, n)
amps = (0.3, 0.5)
colors = ('green', 'blue')

signals = []
 
for f, a, c in zip(freqs, amps, colors): 
    

    y = np.ones(n) * f 
    z = a * np.sin(2 * np.pi * f * x)  + a
    signals.append((x, y, z))
 
    ax.plot(x, y, z, label= f'freq = {f}', color = c)
    
np.sum((xyz[2] for xyz in signals)).shape

ax.plot(x, 
        np.ones(n) * 0,
        np.sum((xyz[2] for xyz in signals)),
        label= f'freq = {f}', 
        color = 'red', linewidth = 3, alpha = 0.6)

plt.ylim(0, 1200)
ax.tick_params(labelbottom = False )
# labelleft', 'labelbottom', 'labelright', 'labeltop'
plt.tight_layout()
plt.show();
```
